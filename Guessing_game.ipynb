```python
import random

colors = ['R', 'B', 'G', 'O', 'P', 'W', 'A', 'K', 'Y']
secret_code = (random.sample(colors, 4))


C = 0
X = 0
Guesses = 7
tries = 0
warnings = 0
looping = True

#print(secret_code)#<--- Undo comment to see the secret code generated by the game.

print("Enter your name: ")
name = input("Enter your name: ")

if name:
    print("Welcome to Color Guesser, " + name)
    print("In this game, a pattern is encrypted as a code of four colored digits as shown in the table below.\n"
          "the objective of this game is to guess the correct code of four digits in 8 attempts. based on your guess, the game will give\n"
          "various hints  such as if a right color is in the correct position or if the color is in the keycode but not in the correct\n"
          "position.  When your guess matches the correct arrangement, you win the game and the game will print the correct answer along\n"
          "with the number of attempts. Good luck!\n")

    print("REMINDERS!\n"
          "1. YOU ONLY GET 8 ATTEMPTS\n"
          "2. Please input the exact corresponding color code that you have chosen\n"
          "failure to put the exact color code or any character not indicated in the\n"
          "color code table below will result in a 'warning'. 3 warnings will result\n"
          "in a forfeit of 1 attempt.\n"
          "3. If you input 2 warnings in a row, it will take a way an additional guess")

    print(
       " -COLOR CODES-\n"
        " ____________\n"
        "| RED    = R |\n"
        "| BLUE   = B |\n"
        "| GREEN  = G |\n"
        "| ORANGE = O |\n"
        "| PURPLE = P |\n"
        "| BROWN  = W |\n"
        "| GRAY   = A |\n"
        "| PINK   = K |\n"
        "| YELLOW = Y |\n"
        "--------------"
          )

while looping:
                player_guess = [x.upper() for x in input("Enter the 4 color code with a space in between: ").split()]
                
                for variable in range(len(player_guess)):
                    try:
                        if  player_guess[variable] == secret_code[variable]:
                            C += 1
                        elif player_guess[variable] != secret_code[variable]:
                            X += 1
                    except IndexError:
                        warnings += 1
                        continue
                    finally:
                        if player_guess[variable]  not in secret_code and player_guess[variable] in colors:
                            X *= 0
                        elif player_guess[variable] not in secret_code and player_guess[variable] not in colors:
                            warnings += 1
                            X *= 0
                if C != 4:
                    print(str(C)+"C")
                    print(str(X)+"X")
                    print(str(Guesses)+" Guesses remaining!")
                    print(str(warnings)+" Warnings")
                    C *= 0
                    X *= 0
                    tries += 1
                    Guesses -= 1
                    print("------------------")
                    if warnings >= 3: 
                       warnings = 0
                       tries -= 1
                       Guesses -=1
                if C >= 4:
                    looping = False
                    print(str(C)+"C")
                    print(str(X)+"X")
                    print(str(tries)+" Guesses!")
                    print("Congratulations!")
                    print("The secret code is: "+ str(secret_code))
                elif tries >= 8 or Guesses <= 0:
                    looping = False
                    print(str(C)+"C")
                    print(str(X)+"X")
                    print(":( you lost!")
                    print("The secret code is: "+ str(secret_code))
            



```
